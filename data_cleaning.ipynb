```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("dataframe_output.csv")
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['claps']=data['claps'].fillna('0')
```

```python
data['claps']=data["claps"].replace({"k":"*1e3","K":"*1e3"}, regex=True).map(pd.eval).astype(float)
```

```python
data['responses']=data['responses'].fillna(0)
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=0,subset=['reading time'],inplace=True)
```

```python
data['reading time'].isnull().sum()
```

```python
data.dropna(axis=0,subset=['title'],inplace=True)
```

```python
calender=data['date']
```

```python
# author name , author pic , image ki ek sheet for making tooltip
tooltip=data.iloc[:,[3,8,9]]
```

```python
data.drop(data.iloc[:,[8,9]],axis=1,inplace=True)
```

```python
data.head()
```

```python
with pd.ExcelWriter('final_project_data.xlsx')as writer:
    data.to_excel(writer,sheet_name="data")
    calender.to_excel(writer,sheet_name="calender")
    tooltip.to_excel(writer,sheet_name="tooltip")
```
